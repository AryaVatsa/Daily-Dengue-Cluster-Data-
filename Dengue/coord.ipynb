```python

import pandas as pd
import geocoder
from math import cos, asin, sqrt

```

```python
localData = pd.read_csv("NeaDataLocalities", index_col= False)
localData
```

```python
placeData = localData["place"]
placeData
```

```python
placeData = placeData.drop_duplicates()
```

```python
placeData
```

```python
uniquePlaces = placeData.reset_index(drop=True)
```

```python
uniquePlaces
```

```python

oldPlaces = pd.read_csv("uniquePlaces")
oldPlaces["place"]
```

```python
oldPlacesList = []
for place in oldPlaces["place"]:
    oldPlacesList.append(place)
oldPlacesList
```

```python
newPlacesList = []
for place in uniquePlaces:
    newPlacesList.append(place)
newPlacesList
```

```python
updateRequiredPlaces = []
updatedLats = []
updatedLongs = []
for item in newPlacesList:
    if item not in oldPlacesList:
        geodata = geocoder.bing(f'{item}, Singapore', key='AmC8YHiucJuxp6YxOjD5ythZxvKMnNX29tEhApaM1Iq8cX0WscnjIrwi8YNKLE0b')
        results = geodata.json
        updatedLats.append(results['lat'])
        updatedLats.append(results['lng'])
        updateRequiredPlaces.append(item)

uniquePlacesData = pd.DataFrame(updateRequiredPlaces)
```

```python
file = open("uniquePlaces", "a")
csv = uniquePlacesData.to_csv(index=False, header=False)
file.writelines(csv)
file.close()
```

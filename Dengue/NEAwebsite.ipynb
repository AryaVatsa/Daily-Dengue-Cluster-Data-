```python
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium import webdriver
# from selenium.webdriver.chrome.service import Service as ChromeService
# from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import datetime
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions


options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-gpu")
driver = webdriver.Chrome(options=options)

```

```python
driver.get("https://www.nea.gov.sg/dengue-zika/dengue/dengue-clusters")
driver.maximize_window()
driver.implicitly_wait(2)
placesRAW = driver.find_elements(By.XPATH, '//tr[@data-row]')
casesSince2weeks = driver.find_elements(By.XPATH, '//td[4]')
casesSinceStart = driver.find_elements(By.XPATH, '//td[5]')

```

```python
places = []
for place in placesRAW:
    places.append(place.get_attribute("id"))
print(places)
```

```python
showButtons = []
for place in places:
    showButtons.append(driver.find_element(By.XPATH, f'//*[@id="{place}"]/td[3]/p[1]'))
def closePopUp():
    driver.find_element(By.XPATH, '//*[@id="ClusterModal"]/div/div[1]/h2/span').click()
localities = []


for button in showButtons:
    button.click()
    driver.implicitly_wait(2)
    elements = driver.find_elements(By.XPATH, '//*[@id="modalBody"]/table/tbody')
    for element in elements:
        localities.append(element.text)
    driver.implicitly_wait(2)
    closePopUp()
localityFile = open("NeaLocalityData", "a")
localityFile.writelines(localities)
localityFile.close()


```

```python
def fileFindnReplace(FileName, Find, Replace):
    file = open(FileName, "r")
    newContent = (file.read()).replace(Find, Replace)
    file.close()
    print(newContent[0])
    Data = open(FileName, "w")
    Data.write(newContent)
    Data.close()



fileFindnReplace("NeaLocalityData", "Location No. of Cases", "")
fileFindnReplace("NeaLocalityData", "\n(", " (")
file = open("NeaLocalityData", "r")
content = file.readlines()
for n, con in enumerate(content):
    if con == "\n":
        content.pop(n)
file.close()
file = open("NeaLocalityData", "w")
print(content)
if "\n" not in content[-1]:
    content[-1] = content[-1] + "\n"
file.writelines(content)
file.close()
```

```python
file = open("NeaLocalityData", "r")
lines = file.readlines()
newLines = []
for line in lines:
    newLine = line
    try:
        if line[-4].isdigit() and line[-3].isdigit() and line[-2].isdigit():
            newLine = line[:-5] + "," + line[-4:]
        
        elif line[-3].isdigit() and line[-2].isdigit():
            newLine = line[:-4] + "," +  line[-3:]

        elif line[-2].isdigit():
            newLine = line[:-3] + "," +  line[-2:]
    except IndexError:
        pass
    newLines.append(newLine)
file = open("NeaLocalityData", "w")
file.writelines(newLines)
file.close()





```

```python
from math import asin, sqrt, cos
def distance(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295
    a = 12742 * asin(sqrt(0.5 - cos((lat2 - lat1) * p)/2 + cos(lat1 * p) * cos(lat2 * p) * (1 - cos((lon2 - lon1) * p)) / 2))
    return round(a,2)
```

```python
wierdPlaces = []
for i, place in enumerate(clusters['place']):
    if clusters.loc[i, "lat"] == 1.35652304 and clusters.loc[i, "long"] == 103.80859375:
        wierdPlaces.append(place)
        clusters.drop(i, axis=0)
clusters.reset_index(drop=True, inplace=True)
wierdPlaces
```

```python
wierdCamps = []
for i, camp in enumerate(SAFCamps['camps']):
    if SAFCamps.loc[i, "lat"] == 1.35652304 and SAFCamps.loc[i, "long"] == 103.80859375:
        wierdCamps.append(camp)
wierdCamps
```

```python


for i, camp in enumerate(SAFCamps["camps"]):
    closestClusterDist = 50
    closestCluster = "need to check"
    for j, NEAcluster in enumerate(clusters["place"]):
        
        dist = distance(SAFCamps.loc[i, 'lat'], SAFCamps.loc[i, 'long'], clusters.loc[j, 'lat'], clusters.loc[j, 'long'])
        if dist < closestClusterDist:
            closestClusterDist = dist
            closestCluster = NEAcluster


    if  closestCluster not in wierdPlaces and camp not in wierdCamps:
        SAFCamps.at[i, 'nearest cluster'] = closestCluster
        SAFCamps.at[i, 'dist betw centre and nearest cluster'] = closestClusterDist
        if SAFCamps.loc[i, 'dist betw centre and nearest cluster'] < 0.1:
            SAFCamps.at[i, 'dist betw centre and nearest cluster'] = 0.1
        if SAFCamps.loc[i, 'dist betw centre and nearest cluster'] < (SAFCamps.loc[i, 'Radius'])/1000:
            SAFCamps.at[i, 'dist from edge < 150?'] = 'YES'
    else:
        SAFCamps.at[i, 'nearest cluster'] = "to be confirmed manually"
        SAFCamps.at[i, 'dist betw centre and nearest cluster'] = 0

SAFCamps = SAFCamps.sort_values(by=['dist from edge < 150?', 'dist betw centre and nearest cluster'])
campsCSV = SAFCamps.to_csv(lineterminator='\n')
file = open("campsNearClusters.csv", "w")
file.write(campsCSV)
file.close()
SAFCamps
```
